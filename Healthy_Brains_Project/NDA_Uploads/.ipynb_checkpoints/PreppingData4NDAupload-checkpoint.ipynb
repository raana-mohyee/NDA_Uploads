```python
# converts interview_age date info to date object
def toDateObj(s):
    m = int(s[0:2])
    d = int(s[3:5])
    y = int(s[6:])
    return dt.date(y, m, d)

# converts interview_date date (or Qualtrics timestamp) info to date object
def toDateObj2(s):
    y = int(s[0:4])
    m = int(s[5:7])
    d = int(s[8:10])
    return dt.date(y, m, d)

# reformat interview_date as MM/DD/YYYY & returns as a series
def reformatIntDate(series):
    series = series.apply(lambda x: toDateObj2(x))
    series = series.apply(lambda x: dt.datetime.strptime(x, "%m/%d/%Y"))
    return series

# a function to convert birthdate to current age in months, returns a series containing the reformatted dates
def calcAgeMonths(m):
    idates = m.forUpload['interview_date'].to_numpy()
    bdays = m.forUpload['interview_age'].apply(lambda x: toDateObj(x))
    bdays = m.forUpload['interview_age'].to_numpy() 
    age_in_mos = pd.Series(np.subtract(idates, bdays))
    return age_in_mos.apply(lambda x: round(x.days/30))
```
